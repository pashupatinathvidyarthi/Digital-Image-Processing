```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1=cv2.imread('baboon.png')
height,width=img1.shape[:2]
quater_height,quater_width=height/4,width/4
T=np.float32([[1,0,quater_width],[0,1,quater_height]])
img_translation=cv2.warpAffine(img1,T,(width,height))
img_translation=cv2.warpAffine(img1,T,(width,height))

fig=plt.figure(figsize=(10,3))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.title('original image')
fig.add_subplot(1,2,2)
plt.imshow(img_translation)
plt.title('translation image')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1=cv2.imread('baboon.png')
height,width=img1.shape[:2]
quater_height,quater_width=height/4,width/4
T=np.float32([[1,0,-quater_width],[0,1,-quater_height]])
img_translation=cv2.warpAffine(img1,T,(width,height))
img_translation=cv2.warpAffine(img1,T,(width,height))

fig=plt.figure(figsize=(10,3))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.title('original image')
fig.add_subplot(1,2,2)
plt.imshow(img_translation)
plt.title('translation image')
plt.show()
```

```python
img2=cv2.imread('baboon.png')
rows,cols=img2.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((cols/2,rows/2), 30, 0.6)
img_rotation=cv2.warpAffine(img2, rotation_matrix, (cols,rows))

plt.imshow(img_rotation)
plt.show()
```

```python
img = cv2.imread('baboon.png')
rows, cols = img.shape[:2]
M = np.float32([[1, 0.5, 0],[0, 1,   0],[0, 0,   1]])
sheared_img = cv2.warpPerspective(img,M,(int(cols * 1.5), int(rows * 1.5)))

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
sheared_rgb = cv2.cvtColor(sheared_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.axis('ON')
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(sheared_rgb)
plt.axis('ON')
plt.title("Sheared Image")

plt.show()
```
