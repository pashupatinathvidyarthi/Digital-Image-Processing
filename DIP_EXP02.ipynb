```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_cameraman=cv2.imread("cameraman.tif")
img_baboon=cv2.imread("baboon.png")
```

```python
plt.imshow(img_cameraman)
img1 = cv2.cvtColor(img_cameraman, cv2.COLOR_BGR2RGB)
plt.imshow(img_cameraman)
plt.axis("off")
plt.show()
```

```python
plt.imshow(img_baboon)
img_baboon = cv2.cvtColor(img_baboon, cv2.COLOR_BGR2RGB)
plt.imshow(img_baboon)
plt.axis("off")
plt.show()
```

```python
print(img_cameraman.shape)
print(img_baboon.shape) #height,width
```

```python
img_cameraman=cv2.resize(img_cameraman,(512,512))
```

```python
img_baboon=cv2.resize(img_baboon,(512,512))
```

```python
plt.imshow(img_cameraman)
plt.axis("off")
plt.show()
```

```python
plt.imshow(img_baboon)
plt.axis("off")
plt.show()
```

```python
#Addition Operation
img_add=cv2.add(img_cameraman,img_baboon)
img_npadd=np.add(img_cameraman,img_baboon)
plt.imshow(img_add)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 4, 1)
plt.imshow(img_cameraman[:,:,::-1])
plt.title('Cameraman')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_baboon[:,:,::-1])
plt.title('Baboon')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_add[:,:,::-1])
plt.title('cv2_Addition')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_npadd[:,:,::-1])
plt.title('np addition')
plt.axis('off')
```

```python
#Substraction Operation
img_sub=cv2.subtract(img_cameraman,img_baboon)
img_npsub=np.subtract(img_cameraman,img_baboon)


plt.imshow(img_sub)
plt.axis("off")
plt.show()
```
