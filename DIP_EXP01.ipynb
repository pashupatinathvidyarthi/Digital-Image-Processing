```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('baboon.png')
plt.imshow(img1)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
plt.imshow(img1[:,:,::-1])
plt.axis('off')
plt.show()
```

```python
img1.shape
H=255
W=255

##resizing of image
img4 = cv2.resize(img1, (100, 100))
img5 = cv2.resize(img1, (50, 50))

fig=plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.imshow(img1[:,:,::-1])
plt.axis("off")
plt.title('Original Image')

fig.add_subplot(1,3,2)
plt.imshow(img4)
plt.axis("off")
plt.title('Size=100x100')

fig.add_subplot(1,3,3)
plt.imshow(img5)
plt.axis("off")
plt.title('Size=50x50')

plt.show()
```

```python
img6=img1[:,:,0]
img7=img1[:,:,1]
img8=img1[:,:,2]

fig=plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.axis("off")
plt.imshow(img6)
plt.title('R_Gray')

fig.add_subplot(1,3,2)
plt.axis("off")
plt.imshow(img7)
plt.title('G_Gray')

fig.add_subplot(1,3,3)
plt.axis("off")
plt.imshow(img8)
plt.title('B_Gray')

plt.show()
```

```python
fig=plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.axis("off")
plt.imshow(img1[:,:,0], cmap='Reds', vmin=0, vmax=255)
plt.title('Red_Channel')

fig.add_subplot(1,3,2)
plt.axis("off")
plt.imshow(img1[:,:,1], cmap='Greens', vmin=0, vmax=255)
plt.title('Green_Channel')

fig.add_subplot(1,3,3)
plt.axis("off")
plt.imshow(img1[:,:,2], cmap='Blues', vmin=0, vmax=255)
plt.title('Blue_Channel')

plt.show()
```
